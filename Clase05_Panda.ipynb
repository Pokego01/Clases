## la funcion .dropna() eliminara las columnas con datos nulos

```python
datosSN=datos.dropna()
datosSN
```

##FUNCIÓN PARA VERIFICAR SI EL GENERO RPG ESTÁ PRESENTE EN LA LISTA



```python
def filtro_rpg(genres):
    return 'RPG' in genres

#Aplica el filtro utilizando apply
datosRPG = datos[datos['Genres'].apply(filtro_rpg)]

# Con esto muestra el dataset en columnas Titulo y Ranting
nombresRPG = datosRPG[['Title', 'Rating']]
nombresRPG.head(20)
```

##Crear un Grafico para los datos de los rpg

```python
datasetRPG = nombresRPG.head(10)
#No es lo optimo debiso a no mostrar los nombres de los juegos
datasetRPG.drop(['Title'],axis=1)['Rating'].plot(kind='bar')
```

```python
datasetRPG = nombresRPG.head(20)
datasetRPG.set_index('Title', inplace=True)  # Cambiar los numeros de rating por los nombres de juegos usando 'Title' como índice
datasetRPG['Rating'].plot(x=None, kind='bar')
```

##Mostrar el prmedio de rating de un genero de videojuegos

```python
ratings_RPG = datosSN[datosSN['Genres'].apply(lambda x: 'RPG' in x)]['Rating']
promedio_RPG = ratings_RPG.mean()

promedio_RPG
```

```python
#Puede aplicarse a cualquier genero
ratings_Indie = datosSN[datosSN['Genres'].apply(lambda x: 'Indie' in x)]['Rating']
promedio_Indie = ratings_Indie.mean()

promedio_Indie
```

##Nuevo DataFrame para comparar sus componentes

```python
Promedios = pd.DataFrame({'Género': ['RPG', 'Indie'], 'Promedio': [promedio_RPG, promedio_Indie]})
Promedios
```

```python
#Convertir lo anterior a Grafico
plt.pie(Promedios['Promedio'], labels = Promedios['Género'], autopct='%1.1f')
# La parte Labels es para que cada color tenga un nombre, mientras que autopct es paraajustar la cantidad de decimales
plt.title('Promedio de Datos por Genero')
##plt.title es para darle un nombre al grafico
plt.show()
### plt.show muestra el grafico obtimizado
```
